Let's make our cat BLUE!

```
output_blue = img.copy()  # The original image is read-only!
output_blue[:, :, :2] = 0
mpimg.imsave("/content/sample_data/bluekitty.webp", output_blue)
plt.imshow(output_blue)
```

We can make our cat GREEN also

```
output_green = img.copy()  # The original image is read-only!
output_green[:, :, :1] = 0
mpimg.imsave("green_kitty.jpg", output_green)
plt.imshow(output_green)
```

And RED - like so,

```
output_red = img.copy()  # The original image is read-only!
output_red[:, :, :0] = 0
mpimg.imsave("red_kitty.jpg", output_red)
plt.imshow(output_red)
```

```
red_kitty = mpimg.imread("red_kitty.jpg")
plt.imshow(red_kitty)
```

```
averages = img.mean(axis=2)  # Take the average of each R, G, and B
mpimg.imsave("bad-gray.jpg", averages, cmap="gray")
```

```
gray_image = mpimg.imread("bad-gray.jpg")
plt.imshow(gray_image)
print(gray_image.shape)
```

```
weights = np.array([0.3, 0.59, 0.11])
#good_gray = img.copy
grayscale = img.copy() @ weights
mpimg.imsave("good-gray.jpg", grayscale, cmap="gray")
```
